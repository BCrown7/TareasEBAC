#### 2) Uso de librería BeautifulSoup y requests.  Parseo de información (e.g. mercadolibre, despegar.com, etc.) 

```python
import requests
from bs4 import BeautifulSoup
import pprint

url = 'https://www.despegar.com.mx/hoteles/'
data = requests.get(url)

headers = {
    'Access-Control-Alow-Origin': '*',
    'Access-Control-Alow-Methods': 'GET',
    'Access-Control-Alow-Headers': 'Content-Type',
    'Access-Control-Max-Age': '3600',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/133.0.0.0 Safari/537.36'
}

response = requests.get(url, headers=headers)

print(response.content[0:1000])
```

```python
url = 'https://www.despegar.com.mx/hoteles/'
data = requests.get(url)
html = BeautifulSoup(data.text, 'lxml')

for item in html.select('.offer-card-container'):
    try:
        print('---------------------------')
        title = item.select('h2')[0].get_text()
        print(title)
    except Exception as e:
        print('')
```
