### Importar la librería NumPy. Generar estadística descriptiva básica en las columnas unit_price y quantity: Cálculo de la media, mediana, moda

```python
# Conversión de columnas a arrays

unit_price_np = df[['Unit price']].to_numpy()
quantity_np = df[['Quantity']].to_numpy()

# Comprobación del tipo de los arrays y sus contenidos

print('Verificación de las clases de los dos arrays:\n')
print(type(quantity_np))
print(type(unit_price_np),'\n')
print('Verificación del contenido de los arrays:\n')
print(unit_price_np[0:10,0])
print(quantity_np[0:10,0])
```

```python
# Estadísticas del array 'Unit Price'

vals1, counts1 = np.unique(unit_price_np, return_counts=True)
moda = vals1[np.argmax(counts1)]
median1 = np.median(unit_price_np)
mean1 = np.mean(unit_price_np)

print('Las estadísticas básicas del Unit Price son las siguientes:\n')
print(f'La moda del array "Unit Price" equivale a {moda:.2f}')
print(f'La mediana del array "Unit Price" corresponde a {median1:.2f}')
print(f'La media del array "Unit Price" sería {mean1:.2f}\n')
print('-' * 60)

# Estadísticas del array 'Quantity'

vals2, counts2 = np.unique(quantity_np, return_counts=True)
moda2 = vals2[np.argmax(counts2)]
median2 = np.median(quantity_np)
mean2 = np.mean(quantity_np)

print('Las estadísticas básicas del Unit Price son las siguientes:\n')
print(f'La moda del array "Quantity" equivale a {moda2:.1f}')
print(f'La mediana del array "Quantity" corresponde a {median2:.1f}')
print(f'La media del array "Quantity" sería {mean2:.1f}')
print('-' * 60)
```

### Tickets promedio

```python
# 1) Ticket promedio por ciudad
print("Ticket promedio por ciudad:\n")
ticket_ciudad = df.groupby('City')['Total'].mean().reset_index()
print(ticket_ciudad, '\n')
print("-" * 50)

# 2) Ticket promedio por product line y ciudad
print("Ticket promedio por product line y ciudad:\n")
ticket_product_city = df.groupby(['Product line', 'City'])['Total'].mean().reset_index()
print(ticket_product_city, '\n')
print("-" * 50)

# 3) Ticket promedio por género
print("Ticket promedio por género:\n")
ticket_gender = df.groupby('Gender')['Total'].mean().reset_index()
print(ticket_gender, '\n')
print("-" * 50)

# 4) Ticket promedio por género y product line
print("\nTicket promedio por género y product line:\n")
ticket_gender_product = df.groupby(['Gender', 'Product line'])['Total'].mean().reset_index()
print(ticket_gender_product, '\n')
print("-" * 50)
```

### Top 5:

```python
def analizar_top_ventas(dataframe, columna_categoria, top_n=6):

    total_sales = dataframe['Total'].sum()

    ventas_por_categoria = dataframe.groupby(columna_categoria)['Total'].sum().reset_index()
    ventas_por_categoria.columns = [columna_categoria, 'Sales']

    ventas_por_categoria['Rank'] = ventas_por_categoria['Sales'].rank(ascending=False) # Calculo de rango

    top_categorias = ventas_por_categoria[ventas_por_categoria['Rank'] < top_n].sort_values(by='Rank') # Top 5

    top_categorias['% Contribution'] = round((top_categorias['Sales'] / total_sales) * 100,2) # Porcentaje de contribución

    return top_categorias

# 1) Top 5 ventas por ciudad
print("Top 5 ventas por ciudad:\n")
top_5_cities = analizar_top_ventas(df, 'City')
print(top_5_cities, '\n')
print("-" * 50)

# 2) Top 5 ventas por tipo de cliente
print("Top 5 ventas por tipo de cliente:\n")
top_5_customer = analizar_top_ventas(df, 'Customer type')
print(top_5_customer, '\n')
print("-" * 50)

# 3) Top 5 ventas por método de pago
print("Top 5 ventas por método de pago:\n")
top_5_payments = analizar_top_ventas(df, 'Payment')
print(top_5_payments, '\n')
print("-" * 50)
```
