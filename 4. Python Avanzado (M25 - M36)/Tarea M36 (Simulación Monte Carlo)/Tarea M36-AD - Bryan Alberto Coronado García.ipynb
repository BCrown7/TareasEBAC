```python
# Importación de librerías

import numpy as np
import matplotlib.pyplot as plt
import random

# 1) La definición de un modelo determinístico de un cálculo de costos: presupuesto simple con ítems, suma uno a uno: METODO TRIANGULAR

# Variables de configuración

num_simulaciones = 5000
max_rolls = 1000
# balance = 100

# Variables/Ítems para presupuesto
costo_a = 0
costo_b = 0
costo_c = 0
costo_d = 0
costo_e = 0
```

```python
# 2) Definición de distribuciones de probabilidad a las variables del modelo

def roll_dice():
        costo_a = np.random.triangular(100,180,260)
        costo_b = np.random.triangular(100,180,260)
        costo_c = np.random.triangular(100,180,260)
        costo_d = np.random.triangular(100,180,260)
        costo_e = np.random.triangular(100,180,260)
        costo_total = np.random.triangular(100,180,260)

        return costo_a, costo_b, costo_c, costo_d, costo_e, costo_total
```

```python
x = roll_dice()

print(x[0], x[1], x[2], x[3], x[4])
```

```python
# 3) Ejecución de simulación aplicada al modelo con 5000 iteraciones

for i in range(num_simulaciones):
    balance = [170]
    num_rolls = [0]

    while num_rolls[-1] < max_rolls:
        iter = roll_dice()

        costo_total = iter[0] + iter[1] + iter[2] + iter[3] + iter[4]
        balance.append(costo_total)
        num_rolls.append(num_rolls[-1] + 1)

# 4) Lectura e interpretación de resultados
# 5) Visualización básica y conclusiones

plt.plot(num_rolls, balance)
plt.xlabel('Prespuestos')
plt.ylabel('Frecuencias')
plt.axvline(np.mean(balance), color='r', linestyle='dashed', linewidth=3)
plt.show()
```

```python
print('Número de muestras tomadas:', balance.__len__())
print('-' * 120)
print(''' Es posible observar que el gráfico nos ayuda a deducir los rangos fijos como estimados, los cuales
      se puede percibir van desde el minimo aproximado de 550, hasta el tope siendo 1100, concentrandose la mayor
      parte de las iteraciones en un rango de 820-900''')
print('-' * 120)
```

```python
# 6) Análisis de sensibilidad con dos tipos de distribución diferentes

print('#1 Gráfico de sensibilidad en distribución triangular')

triangular = np.random.triangular(-5, 0, 5, 5000)
plt.hist(triangular, bins = 100, density = True)
plt.xlabel('Iteraciones')
plt.ylabel('Frencuencias')
plt.show()

print('''   En este análisis es posible verificar la concentración triangular de la distribución, incluso llegando a denotar una similitud
        con la distribución normal, ya que los datos definidos para el gráfico son simétricos, aún así es posible recalcar que
        si genera distintas iteraciones del gráfico, la frecuencia no se concentra en el cero, sino a los laterales de este ''')
```

```python
print('#2 Gráfico de sensibilidad en distribución normal')

normal = np.random.normal(0,1,1000)
plt.hist(normal, bins=100, density=True)
plt.show()

print(''' Para finalizar, con un segundo gráfico proporcionado con distribución normal , podemos comprender la concentración al centro
      de las iteraciones, generando una campana de gauss casi perfecta, lo que idealmente sería un modelo sencillo de distribución''')
```
