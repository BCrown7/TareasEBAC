```python
# Eliminiación de datos nulos para mejor valoración.

df = df.dropna()
df.isnull().sum()
```

```python
# Cambié la columna de año a Integer debido a que resultaba más útil su uso de esta forma que en 'Datetime'

df['Year'] = df['Year'].astype('int32')

df.info()
```

#### Ventas

```python
# Ventas por año (en MDLLS)

df.groupby('Year')['Global_Sales'].sum()
```

```python
# El videojuego más vendido de cada año

df.groupby(['Year','Name'])['Global_Sales'].sum().sort_values(ascending=False)
```

```python
# Plataformas con más ingresos por año

df.groupby(['Year','Platform'])['Global_Sales'].sum().sort_values(ascending=False)
```
