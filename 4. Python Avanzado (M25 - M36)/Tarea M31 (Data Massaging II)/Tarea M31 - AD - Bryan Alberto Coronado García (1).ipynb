```python
df['Nationality'] = df['Nationality'].apply(lambda x: x.title())
df.sample(10)
```

```python
import pytz

df['Zona Horaria'] = df['Nationality'].apply(lambda x: pytz.country_timezones[x])
```

```python
from countryinfo import CountryInfo

df['Capitales'] = df['Nationality'].apply(lambda x: CountryInfo(x).capital())
```
